### Change CUDA version manually
Colab supports Cuda 10.1 by default. Whereas Pedestron requires **10.0/9.2** \\
*  Go to \ folder (parent of /content folder) by clicking at the home dir menu on top of Files section(first icon below upload button). \\

*  In /usr/local, you will find /usr/local/cuda, /usr/local/cuda-10.0, /usr/local/cuda-10.1. The default version is cuda and it refers to usr/local/cuda-10.1 as you can see from usr/local/cuda/version.txt \\

*  Rename usr/local/cuda to usr/local/cudahigher and usr/local/cuda-10.0 to usr/local/cuda
